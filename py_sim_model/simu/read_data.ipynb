```python
def pass_k1():
    k1 = np.empty((0,3),float)
    row = np.array([])
    k1f = open('../kernel1.txt','r')
    line = k1f.readline()
    while line:
        for i in range(3): 
            row = np.append(row,float(line))
            line = k1f.readline()
        k1 = np.append(k1,np.expand_dims(row, axis=0),axis=0)
        row = np.array([])
    k1f.close()
    return k1

def pass_k2():
    k2 = np.empty((0,3),float)
    row = np.array([])
    k2f = open('../kernel2.txt','r')
    line = k2f.readline()
    while line:
        for i in range(3): 
            row = np.append(row,float(line))
            line = k2f.readline()
        k2 = np.append(k2,np.expand_dims(row, axis=0),axis=0)
        row = np.array([])
    k2f.close()
    return k2

def pass_dk1():
    dk1 = np.empty((0,128),float)
    row = np.array([])
    dk1f = open('../dk1.txt','r')
    line = dk1f.readline()
    while line:
        for i in range(128): 
            row = np.append(row,float(line))
            line = dk1f.readline()
        dk1 = np.append(dk1,np.expand_dims(row, axis=0),axis=0)
        row = np.array([])
    dk1f.close()
    return dk1

def pass_dk2():
    dk2 = np.empty((0,10),float)
    row = np.array([])
    dk2f = open('../dk2.txt','r')
    line = dk2f.readline()
    while line:
        for i in range(10): 
            row = np.append(row,float(line))
            line = dk2f.readline()
        dk2 = np.append(dk2,np.expand_dims(row, axis=0),axis=0)
        row = np.array([])
    dk2f.close()
    return dk2

def pass_db1():
    db1 = np.array([])
    with open('../db1.txt') as f:
        for line in f:
            db1 = np.append(db1,float(line))
            if 'str' in line:
                  break
    return db1

def pass_db2():
    db2 = np.array([])
    with open('../db2.txt') as f:
        for line in f:
            db2 = np.append(db2,float(line))
            if 'str' in line:
                  break
    return db2
 

```
